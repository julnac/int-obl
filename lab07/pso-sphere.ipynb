```python
# Import modules
import numpy as np
from matplotlib import pyplot as plt
# Import PySwarms
import pyswarms as ps
from pyswarms.utils.functions import single_obj as fx
from pyswarms.utils.plotters import plot_cost_history

# Set-up hyperparameters
options = {'c1': 0.5, 'c2': 0.3, 'w':0.9}
# c1 – kognitywna część (jak bardzo cząstka ufa własnemu doświadczeniu),
# c2 – społeczna część (jak bardzo cząstka ufa całemu stadu),
# w – współczynnik bezwładności (czyli jak bardzo zachowuje swój kierunek ruchu).

# Call instance of PSO
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options)
# GlobalBestPSO — każda cząstka zna najlepszą pozycję w całym stadzie (globalna strategia).

# Perform optimization
cost, pos = optimizer.optimize(fx.sphere, iters=200)
# fx.sphere — to funkcja celu, ma minimum w punkcie 0,0

# cost — minimalna wartość funkcji znalezionej przez PSO,
# pos — współrzędne punktu, w którym znaleziono to minimum.

# Obtain cost history from optimizer instance
cost_history = optimizer.cost_history

# Plot!
plot_cost_history(cost_history)
plt.show()
```

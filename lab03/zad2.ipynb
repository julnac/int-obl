```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

iris = load_iris()
X = iris.data  # cechy (inputy)
y = iris.target  # klasy (output)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Zbiór treningowy X:\n", X_train)
print("Zbiór treningowy y:\n", y_train)
print("Zbiór testowy X:\n", X_test)
print("Zbiór testowy y:\n", y_test)
```

```python
# Inicjalizacja drzewa decyzyjnego
clf = DecisionTreeClassifier(random_state=42)

# Trenowanie modelu
clf.fit(X_train, y_train)
```

```python
# Wyświetlenie drzewa decyzyjnego w formie graficznej
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Drzewo decyzyjne dla danych irysów")
plt.show()
```

```python
# Predykcja na zbiorze testowym
y_pred = clf.predict(X_test)

# Obliczanie dokładności
accuracy = accuracy_score(y_test, y_pred)
print(f"Dokładność klasyfikatora: {accuracy * 100:.2f}%")
```

```python
# Obliczanie macierzy błędów
cm = confusion_matrix(y_test, y_pred)

# Wyświetlanie macierzy błędów
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Macierz błędów")
plt.show()
```
